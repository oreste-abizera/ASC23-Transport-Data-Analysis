````python
sns.set_style("whitegrid")

# Histogram
plt.figure(figsize=(10,5))
sns.histplot(data=df, x='co2', bins=20, kde=True, color='skyblue')
plt.title("Distribution of Carbon Dioxide Emissions")
plt.xlabel("CO2 Emissions")
plt.ylabel("Frequency")
plt.show()
````

````python
sns.set_style("whitegrid")

# Histogram
plt.figure(figsize=(10,5))
sns.histplot(data=df, x='co2', bins=40, kde=True, color='skyblue')
plt.title("Distribution of Carbon Dioxide Emissions")
plt.xlabel("CO2 Emissions")
plt.ylabel("Frequency")
plt.show()
````

````python
sns.set_style("whitegrid")

# Histogram
plt.figure(figsize=(10,5))
sns.histplot(data=df, x='co2', bins=70, kde=True, color='green')
plt.title("Distribution of Carbon Dioxide Emissions")
plt.xlabel("CO2 Emissions")
plt.ylabel("Frequency")
plt.show()
````

````python
sns.set_style("whitegrid")

# Histogram
plt.figure(figsize=(10,5))
sns.histplot(data=df, x='feScore', bins=30, color='green')
plt.title("FE Score Graph")
plt.xlabel("fe Score")
plt.ylabel("Count")
plt.show()
````

````python
sns.set_style("whitegrid")

# Histogram
plt.figure(figsize=(10,5))
sns.histplot(data=df, x='ghgScore', bins=30, color='green')
plt.title("GHG Score Distribution")
plt.xlabel("ghg Score")
plt.ylabel("Count")
plt.show()
````

**Intepretation of CO2 Diagram**

**Distribution Shape:**

The histogram shows a distribution of CO₂ emissions that is approximately normal, with most vehicles having emissions concentrated in the 290-420 range, forming the peak of the distribution.

**Frequency Analysis:**

Most vehicles have CO₂ emissions between 290 and 420, indicating this as the most common range. Only a small number of vehicles produce extremely low or high emissions, as shown by the low frequencies at the ends of the distribution.

**Bin Size Impact:**

We initially used 20 bins, which made the graph clearer and easier to interpret. However, as the number of bins increased from 20 to 70, the graph became more compressed and harder to read accurately.

**Observations:**

The dataset suggests that the majority of vehicles exhibit moderate CO₂ emissions.

Extreme outliers with unusually high or low emissions are minimal, making the distribution largely unimodal and centered.

````python
# numeric columns only
numeric_data = df.select_dtypes(include=["number"])

# correlation matrix
correlation_matrix = numeric_data.corr()
pallete = sns.diverging_palette(20, 220, as_cmap=True)

# Create the heatmap
plt.figure(figsize=(20, 8))
sns.heatmap(correlation_matrix,  annot=False, cmap=pallete,  cbar=True)
plt.title("Heatmap of Correlation Matrix ")
plt.show()
````

**Interpretation of Heatmap Diagram**

The heatmap shows a correlation matrix that illustrates how different variables in the dataset are related. Below is a detailed interpretation of the findings and suggestions for which columns could be removed if (ghgScore) is the target variable:

**Interpretation:**

**Correlation Analysis:**

**Strong negative correlations with ghgScore:**
Variables like (co2) and (comb) have a strong negative relationship with (ghgScore). This means that as CO₂ emissions and combined fuel consumption increase, the greenhouse gas score tends to decrease.

**Strong positive correlations with ghgScore:**
Variables such as (feScore) (fuel efficiency score) are strongly positively correlated with (ghgScore). This indicates that vehicles with higher fuel efficiency generally receive a higher ghgScore.

**Weak or no correlations:**
Some variables, including (id, year, pv2, and pv4), show little to no correlation with (ghgScore). This suggests they contribute minimally, if at all, to predicting the greenhouse gas score.

**Other significant relationships:**
Variables like (displ) (engine displacement) and (cylinders) may not directly correlate with (ghgScore) but likely influence it indirectly through their impact on fuel consumption variables like (comb and city).

**Columns to Drop if targeting ghgScore:**

* (id): Serves only as an identifier and has no predictive value.
* (year): Shows minimal correlation with (ghgScore) and may not be useful for prediction.
* (pv2 and pv4): Likely represent categorical vehicle types or conditions. Due to their weak correlation with (ghgScore), they can be dropped unless their categorical significance is critical.
* **Highly correlated with other predictors**:
  Variables like (highway, UHighway, city, and UCity) are strongly correlated with (comb). Keeping all of them may lead to multicollinearity. To avoid this, it is advisable to keep only one e.g., comb() and drop the redundant ones.
